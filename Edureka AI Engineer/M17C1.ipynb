```python
#1  Write a function which accepts a CSV file location and split the data randomly in the file into 
# 60% and 40% and save it in two different files namely “First.csv” and “Second.csv”.  Run this function for HouseData.csv 

import pandas as pd
from sklearn.model_selection import train_test_split

def split_data(file):
    df = pd.read_csv(file)
    train, test = train_test_split(df, test_size=0.4)
    train.to_csv("First.csv")
    test.to_csv("Second.csv")
    
split_data("M17_HouseData.csv")
```

```python
#2  Write a function which accepts a CSV and performs Label Encoding for the “Class” column of that file, 
# creates a new Column called “Label” and saves the result in the same file. Run this function for Marketing.csv

from sklearn.preprocessing import LabelEncoder
import pandas as pd

def label_encode(file):
    df = pd.read_csv(file)
    label_encoder = LabelEncoder()
    c = df['class']
    encoded_data = label_encoder.fit_transform(c)
    df['label'] = encoded_data
    df.to_csv("M17_Marketing.csv")
    
label_encode("M17_Marketing.csv")
```
