```python
# this file is taken from Brexit case study and modified to get count of Nouns & Verbs only. 

import nltk
#import docx # needed only for word doc files see M2C2 Brexit file
from nltk.tokenize import word_tokenize as wt
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

def tokenize(content):
    return wt(content)

def NounsCount(doc):
    no_of_tokens = tokenize(doc)
    pos = nltk.pos_tag(no_of_tokens)
    count_noun = 0
    for p in pos:
        if(p[1]=='NN' or p[1]=='NNP' or p[1]=='NNS' or p[1]=='NNPS'):
            count_noun = count_noun + 1
    return(count_noun)

def VerbCount(doc):
    no_of_tokens = tokenize(doc)
    pos = nltk.pos_tag(no_of_tokens)
    count_v = 0
    for p in pos:
        if(p[1]=='VB' or p[1]=='VBD' or p[1]=='VBG' or p[1]=='VBN'or p[1]=='VBP'or p[1]=='VBZ'):
            count_v = count_v + 1
    return(count_v)

doc = open("M3FIFAWorldCup2018.txt", "r")
n_n = NounsCount(doc.read())
doc.close()
doc = open("M3FIFAWorldCup2018.txt", "r") # Read the file again since after the first read, the file has reached its endpoint.
n_v = VerbCount(doc.read())
print("No. of Nouns = "+ str(n_n))
print("No. of Verbs = "+ str(n_v))
doc.close()

y = np.array([n_n, n_v])
my = ["Nouns", "Verbs"]
plt.pie(y, labels = my)
plt.show()
```

```python
# Write a python function which accepts a string and print the first  
# sentence in the string along with its syntax tree. Take function name as “PrintSyntaxTree” 

#nltk.download('punkt')
#nltk.download('averaged_perceptron_tagger')
import nltk
from nltk import pos_tag, word_tokenize as wt, RegexpParser as rp

def PrintSyntaxTree(doc): 
    no_of_tokens = wt(doc)
    pos = nltk.pos_tag(no_of_tokens)
    chunker = rp("""
                       NP: {<DT>?<JJ>*<NN>}    #To extract Noun Phrases
                       P: {<IN>}               #To extract Prepositions
                       V: {<V.*>}              #To extract Verbs
                       PP: {<p> <NP>}          #To extract Prepositional Phrases
                       VP: {<V> <NP|PP>*}      #To extract Verb Phrases
                       """)
    output = chunker.parse(pos)
    print("After Extracting\n", output)
    output.draw() # this will draw in separate window

doc = open("M3FIFAWorldCup2018.txt", "r")
first_line = doc.readline()
print(first_line)
freq = PrintSyntaxTree(first_line)
doc.close()
```

```python
#5 Make a context free grammar having the following rules:  
#a. Noun Phrases are followed by Verb Phrases 
#b. Verb Phrase can have  i. Verb and noun phrase  ii. Verb, noun phrase and a preposition phrase 
#c. Noun Phrases can have  i. Delimiters followed by noun  
#d. Preposition phrase have a preposition followed by a noun phrase 

import nltk
from nltk import pos_tag, word_tokenize as wt, RegexpParser as rp

def PrintSyntaxTree(doc): 
    doc = "She is wearing the beautiful dress, nevertheless how good that is beautiful dress"
    no_of_tokens = wt(doc)
    pos = nltk.pos_tag(no_of_tokens)
    chunker = rp("""
                       NP: {<DT>*<NNP|NNPS>*<VB|VBD|VBG|VBN|VBP|VBZ>*}    #To extract Noun Phrases
                       PP: {<IN>*<NP>*}          #To extract Prepositional Phrases
                       VP: {<V>*<NP|PP>*}      #To extract Verb Phrases
                       """)
    output = chunker.parse(pos)
    print("After Extracting\n", output)
    output.draw() # this will draw in separate window

doc = open("M3FIFAWorldCup2018.txt", "r")
first_line = doc.readline()
print(first_line)
freq = PrintSyntaxTree(first_line)
doc.close()
```
